```python
# üêßüßä revisiting the palmer penguins dataset!
# install the dataset via pip, python package manager
!pip install palmerpenguins
from palmerpenguins import load_penguins_raw

# load dataframe into variable called penguins
penguins = load_penguins_raw()
```

```python
# now ... let's plot flipper length vs. body mass, across the whole dataset
# 1 line of code, make a scatterplot
# flipper length = x-axis, body mass = y-axis

# add your code here

penguins.head()
plt.scatter(penguins['Flipper Length (mm)'], penguins['Body Mass (g)'])
```

```python
# ok ... but now we want to filter the dataset to show 3 species, differentiated

# first make a new, smaller dataframe: all rows, just columns flipper length, body mass, and species

# add code here

flipper_mass = penguins[["Flipper Length (mm)", "Body Mass (g)", "Species"]]
flipper_mass.head()
```

```python
# now drop any NA missing values in this new dataframe

flipper_mass = flipper_mass.dropna()
flipper_mass.head()
```

```python
# NOW make filtered dataframes for each species:
# i.e., an adelie table with only flipper length, body mass, and species
# repeat for gentoo and chinstrap
# so that we can plot each species with different parameters

species = flipper_mass["Species"].unique()
print(species)
# you can also do this by writing out the full name rather than making a list
# but since we learned how to write the list ... this minimizes typos!

adelie = flipper_mass[flipper_mass["Species"] == species[0]]
gentoo = flipper_mass[flipper_mass["Species"] == species[1]]
chinstrap = flipper_mass[flipper_mass["Species"] == species[2]]

# test adelie out, using .head() and .tail()
# are all the rows the same species?

adelie.head()
adelie.tail()
```

```python
plt.figure(figsize=(10, 8))
# setting up for a larger figure

# use color AND marker channels, to show 3 different species
# on the same scatterplot, bill length vs. body mass
# documentation for markers: https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers


# pay close attention to color - is the important information legible?

# add legend, labels, and title


plt.scatter(adelie["Flipper Length (mm)"], adelie["Body Mass (g)"], color = "Orange", marker="o", label = "adelie")
plt.scatter(gentoo["Flipper Length (mm)"], gentoo["Body Mass (g)"], color = "Teal", marker="v", label = "gentoo")
plt.scatter(chinstrap["Flipper Length (mm)"], chinstrap["Body Mass (g)"], color = "Purple", marker="s", label = "chinstrap")

# note: we could loop thru these, or write as 1 line, but we are spelling them out for now

plt.legend()

plt.ylabel("Body Mass (g)")
plt.xlabel("Flipper Length (mm)")
plt.title("\nFlipper Length and Body Mass\nfor 3 Penguin Species, Palmer\n", pad=6, fontsize=18)
```

```python
# what if we wanted to push this in terms of information density, and add ANOTHER attribute?
# what channel would you change (or add?)
# can you keep this from becoming too busy to understand?

# create a figure builds on the previous one
# that visualizes any fourth attribute

# first adding island into the dataframes
flipper_mass = penguins[["Flipper Length (mm)", "Body Mass (g)", "Species", "Island"]]
adelie = flipper_mass[flipper_mass["Species"] == species[0]]
gentoo = flipper_mass[flipper_mass["Species"] == species[1]]
chinstrap = flipper_mass[flipper_mass["Species"] == species[2]]

plt.figure(figsize=(10, 8))

palette = {"Dream": "orange", "Biscoe": "teal", "Torgersen": "purple"}

plt.scatter(adelie["Flipper Length (mm)"], adelie["Body Mass (g)"], c = adelie["Island"].map(palette), marker="o", edgecolor = "black", linewidths=0.5)
plt.scatter(gentoo["Flipper Length (mm)"], gentoo["Body Mass (g)"], c = gentoo["Island"].map(palette), marker="v", edgecolor = "black", linewidths=0.5)
plt.scatter(chinstrap["Flipper Length (mm)"], chinstrap["Body Mass (g)"], c = chinstrap["Island"].map(palette), marker="s", edgecolor = "black", linewidths=0.5)
plt.ylabel("Body Mass (g)")
plt.xlabel("Flipper Length (mm)")
plt.title("\nFlipper Length and Body Mass\nfor 3 Penguin Species on 3 Islands, Palmer\n", pad=6, fontsize=18)

# this makes a shape-only legend, by making empty plots (dummy data)
# kind of hack-y, but the simplest way to illustrate this in matplotlib
plt.scatter([], [], marker="o", color="black", label="Adelie")
plt.scatter([], [], marker="v", color="black", label="Gentoo")
plt.scatter([], [], marker="s", color="black", label="Chinstrap")

# this is just text annotation, directly writing on the plot
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html
plt.text(180, 6300, "Dream", color="orange", fontsize=11)
plt.text(180, 6175, "Biscoe", color="teal", fontsize=11)
plt.text(180, 6050, "Torgersen", color="purple", fontsize=11)

plt.legend()
```

```python
# now: create a HISTOGRAM, which shows distribution of data, with the BODY MASS data for 3 species
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

# on the y-axis, a count of the # of penguins
# on the x-axis, a measure of body mass (g)

# each species should have its own COLOR - choose wisely! (https://matplotlib.org/stable/gallery/color/named_colors.html)
# how do the colors look next to each other? (Albers)

# make sure to add a LEGEND and a title to the histogram

plt.hist(penguins[penguins["Species"] == species[0]]["Body Mass (g)"], color="Orange", alpha = 0.8, label = "Adelie")
plt.hist(penguins[penguins["Species"] == species[1]]["Body Mass (g)"], color="Teal", alpha = 0.8, label = "Gentoo")
plt.hist(penguins[penguins["Species"] == species[2]]["Body Mass (g)"], color="Purple", alpha = 0.8, label = "Chinstrap")

plt.title("Distribution of Body Mass (g) Across 3 Species of Penguins")
plt.xlabel("Body Mass (g)")
plt.ylabel("# of Penguins")
plt.legend()
```

```python
# create a bar chart, with (at least) 2 variables of your choice
# (how will you get the counts for the y-axis?)
# make sure to clean up: label your axes and add a title üé®

islands_only = penguins['Island'].value_counts().reset_index()
islands_only.columns = ['Island', 'Count']
islands_only.head()

# reset_index there makes sure that there is an order 0, 1, 2 to the rows
# which helps me with my bar chart (x = 0 can be Biscoe, x = 1 Dream, etc.)
```

```python
bar_colors = ["Tomato", "Orange", "Khaki"]
plt.bar(islands_only["Island"], islands_only["Count"], color=bar_colors)


plt.title("\nNumber of Penguins in Palmer Penguins Dataset\nAt 3 Different Islands", pad=20)
plt.xlabel("\nIslands in Palmer Archepelago, Antarctica")
plt.ylabel("# of Penguins")
```
